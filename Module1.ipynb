## Intro to NLP
NLP is a field of study in AI, and, as such, has several aspects:
- **Natural Language Understanding (NLU)**: The ability to understand human language, including the meaning of words and sentences.
- **Natural Language Generation (NLG)**: The ability to generate human language, including text and speech.
- **Natural Language Processing (NLP)**: The ability to process human language, including text and speech.

### Applications of NLP:
- Search Engines
- Machine Translation
- Sentiment Analysis
- Chatbots
- Speech Recognition
- Text Summarization
- Named Entity Recognition
- Text Classification
- Question Answering
- News Categorization
- Spam Detection

#### Dataset Types: Structured and Unstructured
-> Structured Dataset: fixed dimensions, well organized, tabular fromat or in key-value pairs, usually excel or sql. Example is table conatining some data or json files containing data.
-> Unstructured Dataset: no fixed dimensions, no structure, not well organized, not in tabular format, usually text files. Example is audio files, video files, images etc i.e A dataset having data of cats and dogs.


#### Unstructured dataset example

*Text  Dataset* : Social media data of posts, tweets, comments; conversation like messages, emails and chats; Articles : news, blogs etc
                - basically words arranged in particular order
                - grammar and defined structures
                - meaning and context

*Speech Dataset* : Audio files of speech, songs, podcasts, audiobooks etc

*Image Dataset* : Images of objects, animals, humans, places etc
